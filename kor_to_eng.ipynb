```
import numpy as np
import pytesseract
from PIL import Image        
import pyttsx3 as txt
from googletrans import Translator     
from google.colab import files 
 
```

```
img = Image.open('test (3).jpg')      
print(img)                           


```

```
result = pytesseract.image_to_string(img, lang = "kor",config='--psm 6')    
```

```
p = Translator()                       

k = p.translate(result,dest='english')       
print(k) 
```
